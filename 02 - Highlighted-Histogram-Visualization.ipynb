This notebook showcases techniques for creating highlighted histograms to emphasize specific data ranges or categories within datasets. It demonstrates interactive and static visualizations, leveraging Python libraries such as Matplotlib and Seaborn, with customizable color schemes, annotations, and threshold-based highlighting. The project includes example datasets, clear methodology for highlighting key insights, and reproducible code to help analysts and researchers effectively communicate distribution patterns and trends.

<h1 style='font-family: Dalton White; font-weight: 600; font-size: 20px; text-align: left'>1.0. Import Required Libraries</h1>

```python
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
import pandas as pd
import numpy as np
import math

import warnings
np.random.seed(1)

warnings.simplefilter("ignore")
pd.set_option('display.max_columns', 10)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print("\n----- Libraries Loaded Successfully -----\n")
```

<h1 style='font-family: Dalton White; font-weight: 600; font-size: 20px; text-align: left'>2.0. Load and Inspect Dataset</h1>

```python
if __name__ == "__main__":
    sample_data = np.random.normal(loc=0.5, scale=0.3, size=500)
    print(sample_data)
```

<h1 style='font-family: Courier New; font-weight: 600; font-size: 20px; text-align: left'>3.0. Histograms Visualization</h1>

<h3 style='font-family: Dubai; font-weight: 600; font-size: 18px; text-align: left'>3.1. Histogram plotting with standard bins</h3>

```python
fig, ax = plt.subplots(figsize=(9,8), dpi=100)
plt.hist(sample_data)
plt.title("Distribution of Electrical Conductivity (dS/m)")
ax.set_xlabel('Electrical Conductivity (dS/m)')
ax.set_ylabel('Frequency/Counts')
plt.show()
```

<h3 style='font-family: Dubai; font-weight: 600; font-size: 18px; text-align: left'>3.2. Customizing Bins, Spines and Grids Styles</h3>

```python
if __name__ == "__main__":
    fig, ax = plt.subplots(figsize=(9,8))
    plt.hist(sample_data, bins=30, edgecolor='k', linewidth=0.6)
    sns.despine(ax=ax, left=True, right=True, bottom=False, top=True)
    plt.title("Distribution of Electrical Conductivity (dS/m)")
    ax.grid(linestyle='--', alpha=0.1, color='black')
    ax.set_xlabel('Electrical Conductivity (dS/m)')
    ax.set_ylabel('Frequency/Counts')
    plt.show()
```

<h3 style='font-family: Dubai; font-weight: 600; font-size: 18px; text-align: left'>3.1. Highlighting Bins Based on its Values</h3>

```python
if __name__ == "__main__":
    fig, ax = plt.subplots(figsize=(9,8))
    sns.despine(ax=ax, left=True, right=True, bottom=False, top=True)
    n, bins, patches = plt.hist(
        sample_data, bins=25,
        facecolor='w',
        edgecolor='k',
        linewidth=0.5,
        alpha=0.7
    )
    
    n = n.astype('int')
    for i in range(len(patches)): patches[i].set_facecolor(plt.cm.viridis(n[i] / max(n)))
    
    plt.title("Distribution of Electrical Conductivity (dS/m)")
    ax.grid(linestyle='--', alpha=0.1, color='black')
    ax.set_xlabel('Electrical Conductivity (dS/m)')
    ax.set_ylabel('Frequency/Counts')
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()
```

<h3 style='font-family: Dubai; font-weight: 600; font-size: 18px; text-align: left'>3.1. Highlighting Specific Bins Based on its Value</h3>

```python
if __name__ == "__main__":
    fig, ax = plt.subplots(figsize=(9,8))
    sns.despine(ax=ax, left=True, right=True, bottom=False, top=True)
    n, bins, patches = plt.hist(
        sample_data, bins=25,
        facecolor='w',
        edgecolor='k',
        linewidth=0.5,
        alpha=0.7
    )
    
    n = n.astype('int')
    for i in range(len(patches)): patches[i].set_facecolor(plt.cm.viridis(n[i] / max(n)))
    
    patches[10].set_fc('red')
    patches[10].set_alpha(1)
    
    plt.title("Distribution of Electrical Conductivity (dS/m)")
    ax.grid(linestyle='--', alpha=0.1, color='black')
    ax.set_xlabel('Electrical Conductivity (dS/m)')
    ax.set_ylabel('Frequency/Counts')
    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()
```

<h3 style='font-family: Dubai; font-weight: 600; font-size: 18px; text-align: left'>3.1. Adding annotations, colors, and overlays for emphasis</h3>
